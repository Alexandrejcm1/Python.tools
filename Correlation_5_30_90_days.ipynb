```python
import pandas as pd
import seaborn as sns
import yfinance as yf
import datetime as dt
import matplotlib.pyplot as plt

symbols = ['EEM','EWW','EWJ','EWG','EWS','EWM','SPY','EWZ','BTC']

# Retrieve the stock data from Yahoo Finance

end = (dt.date.today()).strftime('%Y-%m-%d')
start = (dt.date.today()-dt.timedelta(5)).strftime('%Y-%m-%d')

stock_data = yf.download(symbols, start, end)

stock_prices = stock_data['Close']

correlation_matrix = stock_prices.corr()

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', square=True)

plt.title('Stock Correlation Matrix 5 days')
plt.show()

```
