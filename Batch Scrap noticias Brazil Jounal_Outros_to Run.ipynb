```python
import requests
from bs4 import BeautifulSoup
import time
from IPython.display import clear_output

# depois adicionar mais sites exame, antagonista era token...
#preciso de um screening disso p twitter


def scrape_brazil_journal():
    # Send a GET request to the website
    response = requests.get("https://braziljournal.com/")

    # Parse the HTML content using Beautiful Soup
    soup = BeautifulSoup(response.content, "html.parser")

    # Find the specific headline within the <h1> tags
    headline = soup.find("h1", class_="bannerbrj-title").get_text()

    # Print the headline
    clear_output(wait=True)  # Clear the previous output
    print("Brazil Journal:", headline)
    #print("Batch completed at:", time.ctime())
    

def scrape_Neofeed():
    url = "https://neofeed.com.br/"

# Send a GET request to the URL
    response = requests.get(url)

# Create a BeautifulSoup object with the HTML content
    soup = BeautifulSoup(response.content, "html.parser")

# Find the headline news element
    headline_news = soup.find("div", class_="box-title")

# Extract the text from the headline news element if it exists
    headline_text = headline_news.text.strip() if headline_news else "No headline news found."

# Print the headline news
   
    print("News Neofeed: ", headline_text)
    #print("Batch completed at:", time.ctime())

def scrape_Jardim():
    
    response = requests.get("https://oglobo.globo.com/blogs/lauro-jardim/")

    # Parse the HTML content using Beautiful Soup
    soup = BeautifulSoup(response.content, "html.parser")

    # Find the specific headline within the <h1> tags
    headline = soup.find("h2", class_="feed-post-link").get_text()

    # Print the headline
    
    print("Lauro Jardim: ", headline)
    #print("Batch completed at:", time.ctime())

def scrape_Infomoney():

    # Send a GET request to the website
    response = requests.get("https://www.infomoney.com.br/")

    # Parse the HTML content using Beautiful Soup
    soup = BeautifulSoup(response.content, "html.parser")

    # Find the specific headline within the <h1> tags
    headline = soup.find("span", class_="hl-title hl-title-2").get_text()

    # Print the headline
    
    print("Infomoney:", headline)
    #print("Batch completed at:", time.ctime())
    
def scrape_Valor_investe():

    # Send a GET request to the website
    response = requests.get("https://valorinveste.globo.com/")

    # Parse the HTML content using Beautiful Soup
    soup = BeautifulSoup(response.content, "html.parser")

    # Find the specific headline within the <h1> tags
    headline = soup.find("h2", class_="highlight__title theme-title-element").get_text()

    # Print the headline
   
    print("Valor_investe:", headline)
   

def moneytimes():

    # Send a GET request to the website
    response = requests.get("https://www.moneytimes.com.br/")

    # Parse the HTML content using Beautiful Soup
    soup = BeautifulSoup(response.content, "html.parser")

    # Find the specific headline within the <h1> tags
    headline = soup.find("div", class_="secondary").get_text()

    # Print the headline
   
    print("Moneytimes:", headline)

def blogjota():

    # Send a GET request to the website
    response = requests.get("https://www.jota.info/blog")

    # Parse the HTML content using Beautiful Soup
    soup = BeautifulSoup(response.content, "html.parser")

    # Find the specific headline within the <h1> tags
    headline = soup.find("h2", class_="jota-blog__title").get_text()

    # Print the headline
   
    print("Blog Jota:", headline)   

def poder360():

    # Send a GET request to the website
    response = requests.get("https://www.poder360.com.br/")

    # Parse the HTML content using Beautiful Soup
    soup = BeautifulSoup(response.content, "html.parser")

    # Find the specific headline within the <h1> tags
    headline = soup.find("h2", class_="box-news-list__subhead").get_text()

    # Print the headline
   
    print("Poder 360:", headline)       
    
def BDM():

    # Send a GET request to the website
    response = requests.get("https://www.bomdiamercado.com.br/")

    # Parse the HTML content using Beautiful Soup
    soup = BeautifulSoup(response.content, "html.parser")

    # Find the specific headline within the <h1> tags
    headline = soup.find("a", class_="destak").get_text()

    # Print the headline
   
    print("BDM:", headline)
    
# Run the script in a loop
while True:
    clear_output(wait=True)
    scrape_brazil_journal()
    scrape_Neofeed()
    scrape_Jardim()
    scrape_Infomoney()
    scrape_Valor_investe()
    moneytimes()
    blogjota()
    poder360()
    BDM()
    print("Batch completed at:", time.ctime())
  
    time.sleep(20)  # Wait for 15 seconds
```
